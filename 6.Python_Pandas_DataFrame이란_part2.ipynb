#### boolean selection으로 row 선택하는 방법
- numpy와 동일한 방식으로 해당 조건에 맞는 row만 선택

#### 30대이면서 1등석에 탄 사람만 선택해보기

```python
class_ = td['Pclass'] == 1
age_ = (td['Age'] >= 30) & (td['Age'] < 40)

td_2 = td[class_ & age_]
td_2.iloc[[0,1,2,3],[1,2,3]]
```

### DataFrame에 column 추가, 삭제하는 방법

#### column 추가하는 방법
- [] 사용하여 추가하기
- insert 함수 사용하여 원하는 위치에 추가하기

```python
#broadcasting이 작용
td['Age_double'] = td['Age'] * 2
td.iloc[[0,1,2],[5,12]]
```

```python
td['Age_tripple'] = td['Age'] + td['Age_double']
td = td.rename({'Age_tripple':'Age_sum'},axis= 'columns')
td.loc[[0,1,2],['Age','Age_double','Age_sum']]
```

```python
td.insert(3, 'Fare10', td['Fare']/10)
td.head()
```

#### column 삭제하는 방법
- drop 함수 사용하여 삭제
    - list를 사용하여 멀티플 삭제 가능

```python
#axis사용을 해야하는데 axis = 1로 주어져야한다, 복사본 데이터가 지워지는 것이기 때문에 원본은 남아있다.
#그래서 inplace를 True로 바꿔줘야함
td.drop('Age_sum', axis = 1)
td.head()
```

```python
td.drop(['Age_double','Age_sum'], axis=1, inplace=True)
td.head()
```
